```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2024 - books["published_year"]

```

```python
class CorrelationMap:
    def __init__(self, data: pd.DataFrame):
        self.books = data.copy()
        self.correlation_matrix = None

    def preprocess(self, current_year=2024):
        """Add derived columns for analysis."""
        self.books["missing_description"] = self.books["description"].isna().astype(int)
        self.books["book_age"] = current_year - self.books["published_year"]

    def compute_correlation(self, columns=None, method="spearman"):
        """Calculate the correlation matrix for selected columns."""
        if columns is None:
            columns = ["num_pages", "book_age", "missing_description", "average_rating"]
        self.correlation_matrix = self.books[columns].corr(method=method)

    def plot_heatmap(self):
        """Plot a heatmap of the correlation matrix."""
        if self.correlation_matrix is None:
            raise ValueError("Correlation matrix has not been computed.")
        sns.set_theme(style="white")
        plt.figure(figsize=(8, 6))
        ax = sns.heatmap(
            self.correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
            cbar_kws={"label": "Spearman correlation"}
        )
        ax.set_title("Correlation Heatmap", fontsize=14)
        plt.tight_layout()
        plt.show()
```

```python
HeatMap = CorrelationMap(books)
HeatMap.preprocess()
HeatMap.compute_correlation()
HeatMap.plot_heatmap()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]

book_missing_25_words
```
